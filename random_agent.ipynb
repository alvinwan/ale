```python
%matplotlib inline
import csv
import os

from matplotlib import pyplot as plt
```

```python
LOG_DIR = 'logs'
logs = (
    ('beam_rider', 'log-beam_rider-1482566869.662407.csv'),
    ('breakout', 'log-breakout-1482566910.5739.csv'),
    ('enduro', 'log-enduro-1482567026.239682.csv'),
    ('qbert', 'log-qbert-1482566840.588507.csv'),
    ('seaquest', 'log-seaquest-1482567087.944782.csv'),
    ('space_invaders', 'log-space_invaders-1482567068.470475.csv'))
```

```python
data = {k: [int(line) for line in open(os.path.join(LOG_DIR, v))] for k, v in logs}
```

```python
for title, rewards in data.items():
    averages = []
    for reward in rewards:
        averages.append(
            (sum(averages) + reward)/float(len(averages) + 1))
    
    plt.title(title + (' (%d)' % averages[-1]))
    plt.plot(range(len(averages)), averages)
    plt.figure()
```
